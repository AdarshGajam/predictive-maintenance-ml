```python
# STEP 1: Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

# ----------------------------

# STEP 2: Create / Load Sensor Dataset (Simulated)

np.random.seed(42)

n_samples = 500

temperature = np.random.normal(loc=70, scale=10, size=n_samples)
vibration = np.random.normal(loc=2.0, scale=0.5, size=n_samples)

df = pd.DataFrame({
    "temperature": temperature,
    "vibration": vibration
})

# ----------------------------

# STEP 3: Create Failure Label (Target Variable)

df["failure"] = ((df["temperature"] > 85) | (df["vibration"] > 3)).astype(int)

df.head()
```

```python
import numpy as np
import pandas as pd

np.random.seed(42)

n_samples = 500

runtime_hours = np.linspace(0, 1000, n_samples)

temperature = 40 + 0.03 * runtime_hours + np.random.normal(0, 2, n_samples)
vibration = 1 + 0.002 * runtime_hours + np.random.normal(0, 0.3, n_samples)

df = pd.DataFrame({
    "runtime_hours": runtime_hours,
    "temperature": temperature,
    "vibration": vibration
})

df.head()
```

```python
# Failure happens in last 15% of life + high temp & vibration
runtime_threshold = df["runtime_hours"].quantile(0.85)
temp_threshold = df["temperature"].quantile(0.85)
vib_threshold = df["vibration"].quantile(0.85)

df["failure"] = (
    (df["runtime_hours"] > runtime_threshold) &
    (df["temperature"] > temp_threshold) &
    (df["vibration"] > vib_threshold)
).astype(int)

print("Failure distribution:")
print(df["failure"].value_counts())
```

```python
def rule_based_risk(temp, vib):
    if temp < temp_threshold and vib < vib_threshold:
        return "Low"
    elif temp < temp_threshold or vib < vib_threshold:
        return "Medium"
    else:
        return "High"

df["risk_level"] = df.apply(
    lambda row: rule_based_risk(row["temperature"], row["vibration"]),
    axis=1
)

df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

X = df[["temperature", "vibration"]]
y = df["failure"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.25,
    random_state=42,
    stratify=y
)

print("Training set distribution:")
print(y_train.value_counts())

model = LogisticRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt

plt.scatter(df["temperature"], df["vibration"], c=df["failure"])
plt.xlabel("Temperature")
plt.ylabel("Vibration")
plt.title("Failure Classification")
plt.show()
```

```python
# New sensor readings (example)
new_data = pd.DataFrame({
    "temperature": [65, 75, 90],
    "vibration": [1.8, 2.5, 3.5]
})

# Predict class (0 = no failure, 1 = failure)
predictions = model.predict(new_data)

# Predict probability
probabilities = model.predict_proba(new_data)

results = new_data.copy()
results["Predicted Failure"] = predictions
results["Failure Probability"] = probabilities[:,1]

results
```

```python
def probability_to_risk(p):
    if p < 0.3:
        return "Low Risk"
    elif p < 0.7:
        return "Medium Risk"
    else:
        return "High Risk"

results["Risk Category"] = results["Failure Probability"].apply(probability_to_risk)

results
```
