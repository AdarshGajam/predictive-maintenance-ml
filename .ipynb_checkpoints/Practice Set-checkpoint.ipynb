```python
num = int(input("Enter a number: "))

match num:
    case 1:
        print("Sunday")
    case 2:
        print("Monday")
    case 3:
        print("Tuesday")
    case 4:
        print("Wednesday")
    case 5:
        print("Thursday")
    case 6:
        print("Friday")
    case 7:
        print("Saturday")
    case _:
        print("Enter a number between 1 to 7")
```

```python
a = int(input("Enter a number: "))
b = int(input("Enter a number: "))
operation = input("Enter an operation: ")



match operation:
    case '+':
        print(a + b)
    case '-':
        print(a - b)
    case '*':
        print(a * b)
    case '/':
        print(a / b)
    #case _:
        #print("Enter from these (+, -, *, /) operations")
```

```python
sentence = "Coding in Python is fun"
sum = 0
vowels = ["a","e","i","o","u"]

for char in sentence:
    if(char in vowels):
        sum += 1

print(f"there are {sum} vowels in this sentence")

```
