```python
num = int(input("Enter a number: "))

match num:
    case 1:
        print("Sunday")
    case 2:
        print("Monday")
    case 3:
        print("Tuesday")
    case 4:
        print("Wednesday")
    case 5:
        print("Thursday")
    case 6:
        print("Friday")
    case 7:
        print("Saturday")
    case _:
        print("Enter a number between 1 to 7")
```

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return factorial(n - 1) * n

print(factorial(1))
```

```python
def sum_of_digits(n):
    if n == 0:
        return 0
    
    return n%10 + sum_of_digits(n//10)
# sum of digits of 7532 is same as: 2 (last digit) + sum of digits of 753
print(sum_of_digits(7532))
```
